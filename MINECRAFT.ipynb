The broad question that I will address is: What “kinds” of players are more likely to contribute a large amount of data?
To answer this broad question, I will pose the question:
Which category in the variable experience contributes the most average played hours (AKA data)?
To answer this specific question, I will need the variables in the players dataset, and I will not need to access any data within the sessions dataset, so I will not be wrangling the sessions dataset. Within the players dataset, I will first need to use the group_by() function to aggregate the data and group the unique categories in the experience variable, and use summarize to find the average hours played within each category. Since I am comparing each category by its amount/size, I will be using a bar graph.

```r
#The Player dataset is already tidy since each variable is a single column, each observation is a single row, and each value is in a single
# cell. The variable that is irrelevant to us in this dataset is the hashedEmail column, as it is not helpful in predicting anything, so we will cull
# that column
# I will also change the Data type of Age to Integer, because Age is not generally a decimal.

players_cull <- select(players_data, -hashedEmail) |>
mutate(Age = as.integer(Age))
players_cull
```

```r
#Now that we have our tidy dataset, let's see if the variable experience correlates to more
# played hours

average_experience_hours <- players_cull |>
group_by(experience) |>
summarize(average_hours = round(mean(played_hours, na.rm = TRUE), 2),
         total_hours = sum(played_hours, na.rm = TRUE)) |>
arrange(average_hours)

average_experience_hours
```

```r
#At first glance, it looks like regular players overwhelm the results by a long shot, but first,
# Let's look at the ratio of experienced players to see if the result was due to regular being
# the majority experience of players

experience_number_players <- players_cull |>
group_by(experience) |>
summarize(count = n()) |>
arrange(count)

experience_number_players
```

```r
#From the data frame above, it shows that the average kind of players make up the majority of the dataset
# Every other category is approximately even except Pro and Amateur.
#Let's graph the Experience against average_hours in a bar graph

experience_vs_hours_plot <- average_experience_hours |>
ggplot(aes(x=fct_reorder(experience, average_hours, .desc = TRUE), y=average_hours, fill = experience)) +
geom_bar(stat = "identity") +
labs(x = "Player Experience Level", y = "Average hours played", fill = "Experience Level") +
ggtitle("Relationship between type of player and average hours played")

experience_vs_hours_plot
```
